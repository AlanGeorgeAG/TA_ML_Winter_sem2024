```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
data = pd.read_csv("C:\\Users\\Admin\\Downloads\\company_sales_data.csv")

# Task 1: Line Plot - Total Profit
plt.figure(figsize=(10, 6))
plt.plot(data['month_number'], data['total_profit'], label='Total Profit', color='blue')
plt.title('Total Profit as Line Plot')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend()
plt.grid()
plt.show()
```

```python
# Task 2: Line Plot - Total Units Sold
plt.figure(figsize=(10, 6))
plt.plot(data['month_number'], data['total_units'], linestyle='dashed', color='red', marker='o', linewidth=3, label='Total Units Sold')
plt.title('Total Units Sold')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
# Task 3: Line Plot - Units Sold per Product Type
for column in data.columns[1:7]:  # Assuming columns 1 to 6 are product sales
    plt.plot(data['month_number'], data[column],label=column)
plt.title('Units Sold per Product Type')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend()
plt.grid()
plt.show()
```

```python
# Task 4: Scatter Plot - Toothpaste Sales
plt.figure(figsize=(10, 6))
plt.scatter(data['month_number'], data['toothpaste'], color='green',label='toothpaste sold per month')
plt.title('Toothpaste Sales Scatter Plot')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend()
plt.grid()
plt.show()
```

```python
# Task 5: Bar Chart - Face Wash and Face Cream
plt.figure(figsize=(10, 6))
plt.bar(data['month_number'] + 0.2, data['facewash'], width=0.4, label='Face Wash', color='orange')
plt.bar(data['month_number'] - 0.2, data['facecream'], width=0.4, label='Face Cream', color='blue')
plt.title('Face Wash and Face Cream Sales')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.xticks(data['month_number'])
plt.legend()
plt.grid()
plt.show()
```

```python
# Task 6: Bar Chart - Bathing Soap
plt.figure(figsize=(10, 6))
plt.bar(data['month_number'], data['bathingsoap'], color='purple')
plt.title('Bathing Soap Sales')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.grid()
plt.show()
```

```python
# Task 7: Histogram - Total Profit Distribution
plt.figure(figsize=(10, 6))
plt.hist(data['total_profit'], bins=10,color='skyblue', edgecolor='black')#The bins divide the range of total_profit into intervals
plt.title('Profit Data')
plt.xlabel('Profit Range') # ranges of profits (in dollars)
plt.ylabel('Frequency') #y-axis represents the frequency (count) of months that fall into each profit range.
plt.grid()
plt.show()
```

```python
# Task 8: Pie Chart - Product Sales Distribution
sales = data.iloc[:, 1:7].sum()  # Sum all product sales
#The iloc function is a Pandas method used for selecting data by integer-location indexing (i.e., positions of rows and columns).
plt.figure(figsize=(10, 6))
plt.pie(sales, labels=sales.index, autopct='%1.1f%%')#% indicates that we want to format the text as a percentage.
#The labels for each pie segment are taken from the index of the sales object.
#The first 1 specifies the minimum width , The .1 specifies that 1 decimal place should be shown.
#The f indicates that the number is a floating-point value.
plt.title('Product Sales Distribution')
plt.legend(loc='lower right', bbox_to_anchor=(1.3, -0.1))#bbox_to_anchor defines the exact position of the legend relative to the axes or figure.
#(1.3, -0.1) Moves the legend further to the right and bottom
plt.show()
```

```python
# Task 9: Subplots - Bathing Soap and Face Wash
#axs is a list-like object containing individual axes (subplot) objects.
# Create a figure and two subplots arranged vertically (2 rows, 1 column)
fig, axs = plt.subplots(2, 1, figsize=(10, 12))  # Set figure size to 10x12 for better readability

# Plot data for Bathing Soap Sales on the first subplot (axs[0])
axs[0].plot(data['month_number'], data['bathingsoap'], 
            label='Bathing Soap', color='green')  # Line plot with green color for Bathing Soap
axs[0].set_title('sales of bathingsoap')  # Title for the first subplot
axs[0].set_xlabel('Month Number')  # Label for the x-axis
axs[0].set_ylabel('Units Sold')  # Label for the y-axis
axs[0].legend(loc='upper right')
axs[0].grid()  # Add grid lines to the first subplot for better readability

# Plot data for Face Wash Sales on the second subplot (axs[1])
axs[1].plot(data['month_number'], data['facewash'], 
            label='Face Wash', color='red')  # Line plot with red color for Face Wash
axs[1].set_title('sales of Face Wash')  # Title for the second subplot
axs[1].set_xlabel('Month Number')  # Label for the x-axis
axs[1].set_ylabel('Units Sold')  # Label for the y-axis
axs[1].grid()  # Add grid lines to the second subplot for better readability

# Adjust layout to prevent overlap of subplots (e.g., titles and labels)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Task 10: Stack Plot - Product Sales
plt.figure(figsize=(10, 6))
plt.stackplot(data['month_number'], [data[col] for col in data.columns[1:7]], labels=data.columns[1:7], alpha=0.9)
#alpha Set transparency to 80% for better readability
plt.title('Stack Plot of Product Sales')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend(loc='upper left')
plt.show()
```

```python
# Task 11: Z-Score Normalization
mean = data['total_profit'].mean()
std = data['total_profit'].std()
data['total_profit_normalized'] = (data['total_profit'] - mean) / std
print(f"Mean: {mean}, Standard Deviation: {std}")
```

```python
# Task 12: Logarithmic Transformation
data['total_profit_log'] = np.log(data['total_profit'])
plt.figure(figsize=(10, 6))
plt.plot(data['month_number'], data['total_profit_log'], label='Log Transformed Total Profit')
plt.title('Logarithmic Transformation of Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Log Total Profit')
plt.legend()
plt.grid()
plt.show()
```
