```python
#Aufgabe 1
import csv

# Wörterbuch zur Zuordnung der Ländersuffixe
domain_to_country = {
    "de": "Deutschlan",
    "fr": "Frankreich",
    "us": "Vereinigte Staaten",
    "uk": "Vereinigtes Königreich",
    "es": "Spanien",
    "it": "Italien",
    # Weitere Ländercodes können hier hinzugefügt werden
}

# Funktion zum Extrahieren von Informationen aus E-Mail-Adressen
def extrahiere_kontaktinformationen(email_liste):
    kontakte = []
    
    for email in email_liste:
        # Zerlege die E-Mail-Adresse in Benutzername und Domain
        benutzername, domain = email.split('@')
        
        # Zerlege den Benutzernamen in Vorname und Nachname
        try:
            vorname, nachname = benutzername.split('.')
            vorname, nachname = vorname.capitalize(), nachname.capitalize()
        except ValueError:
            # Wenn der Benutzername nicht im Format vorname.nachname ist
            vorname, nachname = benutzername.capitalize(), ""
        
        # Extrahiere das Ländersuffix aus der Domain
        domain_suffix = domain.split('.')[-1]
        land = domain_to_country.get(domain_suffix, "Unbekannt")
        
        # Füge die Kontaktinformationen zur Liste hinzu
        kontakte.append({
            "Vorname": vorname,
            "Nachname": nachname,
            "Land": land
        })
    
    # Speichere die Kontaktinformationen in einer CSV-Datei
    with open("Kontakte_info.csv", mode="w", newline='', encoding="utf-8") as file:
        writer = csv.DictWriter(file, fieldnames=["Vorname", "Nachname", "Land"])
        writer.writeheader()
        writer.writerows(kontakte)
    
    print("Die Kontaktinformationen wurden erfolgreich in 'Kontakte_info.csv' gespeichert.")

# Beispiel-E-Mail-Liste
email_liste = [
    "max.mustermann@beispiel.de",
    "julia.schmidt@beispiel.fr",
    "john.doe@example.us",
    "lucia.garcia@ejemplo.es",
    "marco.rossi@esempio.it",
    "emma.brown@example.uk"
]

# Funktion aufrufen
extrahiere_kontaktinformationen(email_liste)


# CSV-Datei lesen und Inhalt anzeigen
with open("Kontakte_info.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row) 



```

```python
#Aufgabe 2:

# Funktion zur Analyse der Temperaturen für einen einzelnen Tag
def tagesanalyse(temperaturen):
    max_temp = max(temperaturen)
    min_temp = min(temperaturen)
    durchschnitt_temp = sum(temperaturen) / len(temperaturen)
    bereich = max_temp - min_temp
    
    return {
        "Maximum": max_temp,
        "Minimum": min_temp,
        "Durchschnitt": durchschnitt_temp,
        "Bereich": bereich
    }

# Funktion zur Analyse der Temperaturen für die gesamte Woche
def wochenanalyse(wochentemperaturen):
    wochenstatistik = {
        "Tägliche Statistiken": [],
        "Wochenmaximum": None,
        "Wochenminimum": None,
        "Wochendurchschnitt": 0,
        "Wochenbereich": 0
    }
    
    alle_tagesdurchschnitte = []
    alle_maxima = []
    alle_minima = []

    # Tägliche Statistiken berechnen und zur Wochenstatistik hinzufügen
    for tag, temperaturen in enumerate(wochentemperaturen, start=1):
        tagesstatistik = tagesanalyse(temperaturen)
        wochenstatistik["Tägliche Statistiken"].append((f"Tag {tag}", tagesstatistik))
        
        # Werte für die Wochenstatistik sammeln
        alle_tagesdurchschnitte.append(tagesstatistik["Durchschnitt"])
        alle_maxima.append(tagesstatistik["Maximum"])
        alle_minima.append(tagesstatistik["Minimum"])

    # Wochenstatistik berechnen
    wochenstatistik["Wochenmaximum"] = max(alle_maxima)
    wochenstatistik["Wochenminimum"] = min(alle_minima)
    wochenstatistik["Wochendurchschnitt"] = sum(alle_tagesdurchschnitte) / len(alle_tagesdurchschnitte)
    wochenstatistik["Wochenbereich"] = wochenstatistik["Wochenmaximum"] - wochenstatistik["Wochenminimum"]
    
    return wochenstatistik

# Beispielhafte Temperaturdaten (jede Liste repräsentiert einen Tag mit 24 Stunden)
wochentemperaturen = [
    [12, 14, 15, 14, 13, 12, 11, 12, 14, 16, 18, 20, 21, 23, 22, 20, 18, 16, 15, 14, 13, 12, 11, 10],
    [10, 12, 13, 12, 11, 10, 9, 10, 11, 13, 15, 17, 18, 19, 18, 17, 16, 14, 13, 12, 11, 10, 9, 8],
    [8, 10, 11, 12, 11, 10, 9, 10, 12, 14, 16, 18, 19, 21, 20, 18, 16, 14, 13, 12, 11, 10, 9, 8],
    [7, 9, 10, 11, 10, 9, 8, 9, 10, 12, 14, 16, 17, 19, 18, 17, 16, 15, 14, 13, 12, 11, 10, 9],
    [6, 8, 9, 10, 9, 8, 7, 8, 9, 11, 13, 15, 16, 17, 16, 15, 14, 13, 12, 11, 10, 9, 8, 7],
    [8, 10, 11, 12, 11, 10, 9, 10, 11, 13, 15, 17, 18, 19, 18, 17, 16, 15, 14, 13, 12, 11, 10, 9],
    [9, 11, 12, 13, 12, 11, 10, 11, 12, 14, 16, 18, 19, 20, 19, 18, 17, 16, 15, 14, 13, 12, 11, 10]
]

# Wochenanalyse aufrufen und Ergebnisse anzeigen
ergebnis = wochenanalyse(wochentemperaturen)
print("Wochenstatistik:")
for tag, statistik in ergebnis["Tägliche Statistiken"]:
    print(f"{tag}: {statistik}")
print("\nGesamtwochenstatistik:")
print(f"Wochenmaximum: {ergebnis['Wochenmaximum']}")
print(f"Wochenminimum: {ergebnis['Wochenminimum']}")
print(f"Wochendurchschnitt: {ergebnis['Wochendurchschnitt']:.2f}")
print(f"Wochenbereich: {ergebnis['Wochenbereich']}")
```

```python
# Aufgabe 3:

# Notenbuchsystem
class Gradebook:
    def __init__(self):
        # Wörterbuch zur Speicherung von Schülern und Noten
        self.schueler_noten = {}

    def schueler_hinzufuegen(self, schueler_name):
        # Fügt einen neuen Schüler mit leerem Notenverzeichnis hinzu
        if schueler_name not in self.schueler_noten:
            self.schueler_noten[schueler_name] = {}
            print(f"{schueler_name} wurde zum Notenbuch hinzugefügt.")
        else:
            print(f"{schueler_name} ist bereits im Notenbuch vorhanden.")

    def note_hinzufuegen(self, schueler_name, fach, note):
        # Fügt die Note für ein bestimmtes Fach hinzu
        if schueler_name in self.schueler_noten:
            if fach not in self.schueler_noten[schueler_name]:
                self.schueler_noten[schueler_name][fach] = []
            self.schueler_noten[schueler_name][fach].append(note)
            print(f"Note {note} in {fach} für {schueler_name} hinzugefügt.")
        else:
            print(f"{schueler_name} ist nicht im Notenbuch vorhanden.")

    def durchschnitt_berechnen(self, schueler_name):
        # Berechnet den Durchschnitt der Noten eines Schülers über alle Fächer
        if schueler_name in self.schueler_noten:
            gesamt_summe = 0
            anzahl_noten = 0
            for fach, noten in self.schueler_noten[schueler_name].items():
                gesamt_summe += sum(noten)
                anzahl_noten += len(noten)
            if anzahl_noten > 0:
                durchschnitt = gesamt_summe / anzahl_noten
                print(f"Durchschnittsnote für {schueler_name}: {durchschnitt:.2f}")
                return durchschnitt
            else:
                print(f"{schueler_name} hat noch keine Noten.")
        else:
            print(f"{schueler_name} ist nicht im Notenbuch vorhanden.")

    def hoechste_niedrigste_note(self, fach):
        # Bestimmt die höchste und niedrigste Note in einem bestimmten Fach
        hoechste_note = None
        niedrigste_note = None
        for noten in [self.schueler_noten[schueler][fach] for schueler in self.schueler_noten if fach in self.schueler_noten[schueler]]:
            if noten:
                max_note = max(noten)
                min_note = min(noten)
                if hoechste_note is None or max_note > hoechste_note:
                    hoechste_note = max_note
                if niedrigste_note is None or min_note < niedrigste_note:
                    niedrigste_note = min_note
        if hoechste_note is not None and niedrigste_note is not None:
            print(f"Höchste Note in {fach}: {hoechste_note}")
            print(f"Niedrigste Note in {fach}: {niedrigste_note}")
            return hoechste_note, niedrigste_note
        else:
            print(f"Keine Noten in {fach} vorhanden.")
            return None, None

# Beispiel zur Nutzung des Notenbuchsystems
gradebook = Gradebook()

# Schüler hinzufügen
gradebook.schueler_hinzufuegen("Anna Müller")
gradebook.schueler_hinzufuegen("Max Mustermann")

# Noten hinzufügen
gradebook.note_hinzufuegen("Anna Müller", "Mathematik", 2)
gradebook.note_hinzufuegen("Anna Müller", "Mathematik", 4)
gradebook.note_hinzufuegen("Max Mustermann", "Physik", 3)
gradebook.note_hinzufuegen("Max Mustermann", "Physik", 1)

# Durchschnittsnote berechnen
gradebook.durchschnitt_berechnen("Anna Müller")
gradebook.durchschnitt_berechnen("Max Mustermann")

# Höchste und niedrigste Note in einem Fach ermitteln
gradebook.hoechste_niedrigste_note("Mathematik")
gradebook.hoechste_niedrigste_note("Physik")
```
